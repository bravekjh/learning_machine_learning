```python
# %matplotlib notebook
%matplotlib inline
import numpy as np
import sklearn
from sklearn.datasets import load_boston
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import math

matplotlib.rcParams['figure.figsize'] = (30,30)
```

```python
def create_dataframe(data_input):
    if callable(data_input):
        data = data_input()
        df = pd.DataFrame(data.data)
        df.columns = data.feature_names
        df['y'] = data.target
        return df
```

```python
bos = create_dataframe(load_boston)
bos.head()
```

```python
def forward_prop(x, w, bias):
    return x.dot(w) + bias
```

```python
def init_weights(features):
    w = np.zeros(features)
    b = np.zeros(1)
    return w, b
```

```python
def cost(y_hat, y):
    return 0.5 / y.shape[0] * np.sum(np.square(y_hat - y))
```

```python
def compute_gradient(y_hat, y, x):
    diff = y_hat - y
    delta_w = diff.T.dot(x) / y.shape[0]
    delta_b = np.sum(diff) / y.shape[0]
    return delta_w, delta_b
```

```python
def train(lr, iteration, x, y):
    w, b = init_weights(x.shape[1])
    print_freq = iteration / 10 if iteration / 10 >= 1 else 1 
    for i in xrange(iteration):
        y_hat = forward_prop(x ,w,b)
        cst = cost(y_hat, y)
        d_w, d_b = compute_gradient(y_hat, y, x)
        if i % print_freq == 0:
            print "%s. cost: %s " % (i, cst)
#             print "%s. cost: %s w: %s b: %s d_w: %s d_b: %s" % (i, cst, w, b, d_w, d_b)
        w -= lr * d_w
        b -= lr * d_b
    return w, b
```

```python
def normal_equation(x, y, lmbd):
    m = x.shape[0]
    n = x.shape[1] + 1
    bias = np.ones(m)
    x = np.column_stack((x, bias))
    xT_x = x.T.dot(x) 
    L = np.identity(n)
    L[0] = 0
    lmbd_L = lmbd * L
    xT_x_lmbd_L = xT_x + lmbd_L
    inv_xT_x_lmbd_L = np.linalg.inv(xT_x_lmbd_L)
    xt_y = x.T.dot(y)
    theta = inv_xT_x_lmbd_L.dot(xt_y)
    return theta
```
